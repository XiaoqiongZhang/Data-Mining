```python
import pandas as pd
import pandas_datareader.data as web
import datetime as dt
```

```python
start = dt.datetime(2015,1,1)
end = dt.datetime(2019,6,30)
```

```python
stocks = ['AAPL','CVX','IBM','XOM','GS','BA','TEVA','CME']

data = pd.DataFrame()
for i in stocks:
    stock = web.get_data_yahoo(i,start,end)
    data[i] = stock['Adj Close']
```

```python
data.head()
```

```python
data.to_csv('FinanceData.csv')
```
